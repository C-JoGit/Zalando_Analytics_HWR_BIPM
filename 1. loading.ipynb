```python
# filter files only for our project, any language
all_files = os.listdir(path='twitterdata')
r = re.compile(".*_(@zalando|zalando|boohoo|bohoo|asos)")
files = list(filter(r.match, all_files)) 
```

```python
files_split = [file.split('_') for file in files]
files_split

#we should fill missing values for keyword and language based on the filename. Can do that later
```

```python
#make one dataframe out of all these files
dfs = []

files_paths = ['twitterdata/' + file for file in files]

for x in files_paths:
    data = pd.read_json(x)
    dfs.append(data)

df = pd.concat(dfs)
```
